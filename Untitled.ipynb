```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from pandas.plotting import scatter_matrix
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
def data_pre(df):
    df['HS_N'] = 0
    df['HS'] = df['HS'].astype('str')
    for i in range(df.shape[0]):
        if (df['HS'][i])[-1]=="*": 
            df['HS'][i]=(df['HS'][i])[:-1]
            df['HS_N'][i]=1
    df['HS'] = df['HS'].astype('int64')
    
    
    for i in range(df.shape[0]):
        if df['Avg'][i]=="-": 
            df['Avg'][i]=0
    df['Avg'] = df['Avg'].astype('float64')
    
    df['Runs_from_boundary'] = 4*df['4s'] + 6*df['6s']
    df['Runs_from_running'] = df[df.columns[4]] - df['Runs_from_boundary']
    df['Runs_per_Match'] = df[df.columns[4]]/df['Mat']
    df['Runs_per_Innings'] = df[df.columns[4]]/df['Inns']
    df['4s_per_Match'] = df['4s']/df['Mat']
    df['4s_per_Innings'] = df['4s']/df['Inns']
    df['6s_per_Match'] = df['6s']/df['Mat']
    df['6s_per_Innings'] = df['6s']/df['Inns']
    df['50+_score'] = df[100] + df[50]
    df['50+_per_Match'] = df['50+_score']/df['Mat']
    df['50+_per_Innings'] = df['50+_score']/df['Inns']
    df['number_of_outs'] = df['Inns'] - df['NO']
    df['batting_per_match'] = df['Inns']/df['Mat']
    df['SR_without_bounadry'] = df['Runs_from_running']/(df['BF'] - df['4s'] - df['6s']+1)
    df['6_to_4_ration'] = df['6s']/(df['4s']+1)
    df['50_to_100_conversion_rate'] = df[100]/(df['50+_score']+1)
    df['Atleast_one_50+_score'] = df['HS']>=50
    df['Atleast_one_50+_score'] = df['Atleast_one_50+_score'].astype('int64')
    
    #for i in range(10):
    #    df[df.columns[4]+str(i)] = df[df.columns[4]]
    
    return df
```

```python
df = data_pre(df)
```

```python
from sklearn.model_selection import train_test_split
X = df.drop(['2019_Runs','PLAYER'], axis=1)
y = df['2019_Runs']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import StandardScaler
scaler_x = StandardScaler()
scaler_y = StandardScaler()
X = scaler_x.fit_transform(X)
y = scaler_y.fit_transform(np.array(y, dtype='float').reshape((-1,1)))
```

```python
X.max(), X.min(), y.max(), y.min()
```

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import make_scorer
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import Ridge
from xgboost import XGBRegressor
from sklearn.ensemble import ExtraTreesRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.ensemble import AdaBoostRegressor

models = []
models.append(('LIR', LinearRegression()))
models.append(('RID', Ridge()))
models.append(('RFR', RandomForestRegressor()))
models.append(('XGB', XGBRegressor()))
models.append(('ETR', ExtraTreesRegressor()))
models.append(('BAR', BaggingRegressor()))
models.append(('ABR', AdaBoostRegressor()))
# evaluate each model in turn
results = []
names = []
for name, model in models:
    kfold = KFold(n_splits=10, random_state=1, shuffle=True)
    cv_results = cross_val_score(model, X, y.ravel(), cv=kfold, scoring=make_scorer(mean_squared_error))
    results.append(cv_results*1000)
    names.append(name)
    print('%s: %f (%f)' % (name, cv_results.mean()*1000, cv_results.std()*1000))
```

```python
plt.boxplot(results, labels=names)
plt.title('Algorithm Comparison')
plt.show()
```

```python
from sklearn.ensemble import StackingRegressor
reg = StackingRegressor( estimators=models, final_estimator=Ridge())
kfold = KFold(n_splits=10, random_state=1, shuffle=True)
cv_results = cross_val_score(reg, X, y.ravel(), cv=kfold, scoring=make_scorer(mean_squared_error))
print('%s: %f (%f)' % ('Stacking', cv_results.mean()*1000, cv_results.std()*1000))
```

```python
from sklearn.ensemble import VotingRegressor
vot = VotingRegressor(models)
kfold = KFold(n_splits=10, random_state=1, shuffle=True)
cv_results = cross_val_score(vot, X, y.ravel(), cv=kfold, scoring=make_scorer(mean_squared_error))
print('%s: %f (%f)' % ('Voting', cv_results.mean()*1000, cv_results.std()*1000))
```
